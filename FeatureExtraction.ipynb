### Make a new column for episode count

```python
df.loc[4]["Title"]
```

```python
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i == "(":
            check = True
        if check:
            data += i
        if i == ")" and check:
            break
    return data
```

```python
df["Episodes"] = df["Title"].apply(extract_episodes)
```

```python
df["Episodes"] = df["Episodes"].str.replace(" eps", "")
```

```python
df["Episodes"] = df["Episodes"].str.strip("()").astype(int)
```

### Make New column for time stamp

```python
df.loc[0]["Title"]
```

```python
def extract_time(txt):
    start = txt.find(")")
    if start != -1:
        # Extract the substring right after ")"
        sub_txt = txt[start+1:]
        # Now extract the date portion (typically looks like 'Apr 2009 - Jul 2010')
        import re
        match = re.search(r"[A-Za-z]{3} \d{4} - [A-Za-z]{3} \d{4}", sub_txt)
        if match:
            return match.group()
    return ""
```

```python
df["Total Time"] = df['Title'].apply(extract_time)
```

```python
df.head()
```

```python
from datetime import datetime

def calculate_total_months(time_str):
    try:
        # Step 1: Split the input string
        start_str, end_str = time_str.split(" - ")

        # Step 2: Convert both parts to datetime objects
        start_date = datetime.strptime(start_str, "%b %Y")  # "Apr 2009"
        end_date = datetime.strptime(end_str, "%b %Y")      # "Jul 2010"

        # Step 3: Calculate total months difference
        total_months = (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month)

        # Include both start and end months
        return total_months + 1

    except Exception as e:
        return f"Error: {e}"
```

```python
df["Months"] = df['Total Time'].apply(calculate_total_months)
```
